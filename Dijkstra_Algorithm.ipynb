```python
# given data
heights = [[1,2,1,1,1],[1,2,1,2,1],[1,2,1,2,1],[1,2,1,2,1],[1,1,1,2,1]]
# get row and col
row = len(heights)
col = len(heights[0])
# S set (visited Node) and U set (unvisited neighbor Node)
S = {}
U = {}
i,j = 0,0
dist = 0

# a function that update distances 
def update(Node,val):
    i,j = Node
    if i>=0 and i<row and j>=0 and j<col:
        # vaild Node
        if Node not in S:
            Node_val = heights[i][j]
            #print(Node,Node_val)
            d = max(abs(Node_val - val), dist)
            #print(d)
            if Node in U:
                U[Node] = min(U[Node],d)
            else:
                U[Node] = d
            

# repeat until to the end Node
while i<row and j<col:
    # put Node (i,j) into S
    S[(i,j)] = dist
    val = heights[i][j]
    print('S',S)
    # check if traget Node in S
    if (row-1,col-1) in S:
        print('found,the result should be:',S[row-1,col-1])
        break
    # neighbor
    l = (i,j-1)
    r = (i,j+1)
    u = (i-1,j)
    d = (i+1,j)
    # update neighbor node distances
    update(l,val)
    update(r,val)
    update(u,val)
    update(d,val)
    print('U',U)
    # pick the node that has min dist into S
    dist = min(U.values())
    for k,v in U.items():
        if v==dist:
            i,j = k
            break
    
    # remove this node from U
    U.pop((i,j))
    print(U)
    print(dist)
    print((i,j))
```
